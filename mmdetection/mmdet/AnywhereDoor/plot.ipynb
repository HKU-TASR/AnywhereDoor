```python
import numpy as np
import matplotlib.pyplot as plt
from math import pi

VOC0712_TT = [97.6, 83.0, 98.2, 78.5, 92.0, 73.0]
VOC0712_TF = [89.7, 89.2, 51.6, 1.6, 82.7, 71.7]
VOC0712_FT = [96.4, 54.3, 97.9, 11.3, 92.6, 72.0]
VOC0712_FF = [16.7, 42.7, 50.2, 0.4, 49.6, 69.1]

labels = ['Untargeted Remove ASR', 'Targeted Remove ASR', 'Untargeted Misclassify ASR', 'Targeted Misclassify ASR', 'Untargeted Generate ASR', 'Clean mAP']
num_vars = len(labels)

# Figure VOC0712
fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(polar=True))
# 设置刻度0-100
plt.ylim(0, 100)

angles = [n / float(num_vars) * 2 * pi for n in range(num_vars)]
angles += angles[:1]

VOC0712_TT += VOC0712_TT[:1]
ax.plot(angles, VOC0712_TT, linewidth=1, linestyle='solid', label='AnywhereDoor', color='b')
ax.fill(angles, VOC0712_TT, 'b', alpha=0.1)
VOC0712_TF += VOC0712_TF[:1]
ax.plot(angles, VOC0712_TF, linewidth=1, linestyle='solid', label='Baseline + Stratified Sampling', color='g')
ax.fill(angles, VOC0712_TF, 'g', alpha=0.1)
VOC0712_FT += VOC0712_FT[:1]
ax.plot(angles, VOC0712_FT, linewidth=1, linestyle='solid', label='Baseline + Disentanglement', color='r')
ax.fill(angles, VOC0712_FT, 'r', alpha=0.1)
VOC0712_FF += VOC0712_FF[:1]
ax.plot(angles, VOC0712_FF, linewidth=1, linestyle='solid', label='Baseline', color='y')
ax.fill(angles, VOC0712_FF, 'y', alpha=0.1)

plt.xticks(angles[:-1], labels)
ax.legend(loc='upper right', bbox_to_anchor=(0.1, 0.1))
```
